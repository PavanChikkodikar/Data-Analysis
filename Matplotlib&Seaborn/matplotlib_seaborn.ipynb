**Seaborn**


```python
plt.figure(figsize=(10, 6))
sns.histplot(sales['Unit_Cost'], bins=20, kde=True)
plt.title('Histogram of Unit Cost')
plt.xlabel('Unit Cost')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x=sales['Unit_Cost'], orient='h')
plt.title('Box Plot of Unit Cost')
plt.xlabel('Unit Cost')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=sales['Unit_Cost'], y=sales['Revenue'])
plt.title('Scatter Plot of Unit Cost vs Revenue')
plt.xlabel('Unit Cost')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales, x=sales.index, y='Unit_Cost')
plt.title('Line Plot of Unit Cost over Time')
plt.xlabel('Index')
plt.ylabel('Unit Cost')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales, x=sales.index, y='Unit_Cost', ci=None)
plt.fill_between(sales.index, sales['Unit_Cost'], alpha=0.3)
plt.title('Area Plot of Unit Cost over Time')
plt.xlabel('Index')
plt.ylabel('Unit Cost')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.kdeplot(sales['Unit_Cost'], fill=True)
plt.title('Kernel Density Estimation (KDE) Plot of Unit Cost')
plt.xlabel('Unit Cost')
plt.ylabel('Density')
plt.show()
```
